# Step 1: Setting Up Your Environment
Let's apply t-SNE on a high-dimensional dataset. First, let's import the data.


```python
# Import necessary libraries
from sklearn.datasets import make_blobs
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Generate synthetic dataset
X, y = make_blobs(n_samples=300, centers=3, n_features=10, random_state=42)
```

# Step 2: Applying t-SNE

```python
# Apply t-SNE
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)
```

# Step 3: Visualizing the t-SNE result

```python
# Plot t-SNE result
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='tab10')
plt.title('t-SNE Visualization of High-Dimensional Data')
plt.colorbar(scatter)
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.show()
```
