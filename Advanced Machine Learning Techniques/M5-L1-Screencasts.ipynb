## M5L1SC3: Getting Started with Auto-sklearn for Model Selection

### Step 1: Preparing Your Environment
Install Auto-sklearn and essential libraries for setting up your project.

```python
!pip install auto-sklearn
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import autosklearn.classification
```

### Step 2: Loading and Splitting the Dataset
Load the Iris dataset and split it into training and testing sets.

```python
# Load the Iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### Step 3: Automating Model Selection with Auto-sklearn
Initialize and run Auto-sklearn to find and configure the best models for the Iris dataset.

```python
# Initialize the Auto-sklearn classifier
automl = autosklearn.classification.AutoSklearnClassifier(time_left_for_this_task=60,
                                                          per_run_time_limit=30, n_jobs=-1)

# Fit the classifier with the training data
automl.fit(X_train, y_train)

# Print the leaderboard
print(automl.leaderboard())
```

### Step 4: Evaluating Model Performance
Evaluate the best model's performance using accuracy or another performance metric on the test data.

```python
# Make predictions and evaluate
predictions = automl.predict(X_test)
accuracy = (predictions == y_test).mean()
print(f"Accuracy: {accuracy:.2f}")
```

## M5L1SC4: Feature Engineering and Pipeline Analysis with Auto-sklearn

### Step 1: Setting Up Your Auto-sklearn Environment
Ensure Auto-sklearn and other relevant libraries are installed and ready.

```python
# !pip install auto-sklearn
import os
os.system('pip install auto-sklearn > /dev/null 2>&1')

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
import autosklearn.classification
```

### Step 2: Preparing Your Dataset
Action: Load and split the dataset for model training and validation.

```python
# Load the Wine dataset
wine = load_wine()
X, y = wine.data, wine.target

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### Step 3: Automating Feature Engineering with Auto-sklearn
Implement Auto-sklearn to automate feature engineering and pipeline generation.

```python
# Initialize the Auto-sklearn classifier
automl = autosklearn.classification.AutoSklearnClassifier(time_left_for_this_task=120,
                                                          per_run_time_limit=30)

# Fit the model with training data
automl.fit(X_train, y_train)
```

### Step 4: Analyzing the Pipeline Components

Inspect the pipeline details generated by Auto-sklearn for insight validation.

```python
# Retrieve pipeline components
for model in automl.leaderboard()['model_id']:
    pipeline = automl.get_models_with_weights()[model].steps
    print(pipeline)
```

### Step 5: Evaluating and Expanding Pipelines
Evaluate models and suggest enhancements for deeper exploration.

```python
# Evaluate the best model
predictions = automl.predict(X_test)
accuracy = (predictions == y_test).mean()
print(f"Test Set Accuracy: {accuracy:.2f}")

# Explore advanced configurations
print("Consider extending runtime or preprocessor complexity for enhanced performance.")
```
