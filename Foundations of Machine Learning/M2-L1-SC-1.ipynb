# Unsupervised Learning Techniques

Unsupervised learning dives into data without labels, finding hidden patterns or structures.

- **Clustering** groups similar data points into clusters, like K-Means or hierarchical clustering with its dendrogram cut points, tackling tasks like market segmentation or anomaly detection.
- **Dimensionality Reduction** shrinks data dimensions while keeping its core, using PCA or t-SNE, solving visualization and anomaly detection challenges.

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Create function for Netflix clustering example
def netflix_clustering_example():
    # Create synthetic user viewing data
    # Each row represents a user, columns represent features:
    # [action_movies_watched, comedy_movies_watched]
    user_data = np.array([
        [5, 7],    # Moderate action, more comedy
        [1, 2],    # Little of both
        [8, 8],    # Lots of both
        [12, 10],  # Many action movies
        [2, 3],    # Little of both
        [4, 6],    # Moderate action, comedy
        [10, 9],   # High action, high comedy
        [3, 1],    # Low action, very low comedy
        [7, 5],    # Moderate action, moderate comedy
        [11, 12],  # High action, high comedy
        [2, 8],    # Low action, high comedy
        [9, 3],    # High action, low comedy
        [6, 4],    # Moderate action, low comedy
        [1, 5],    # Very low action, moderate comedy
        [13, 11],  # Very high action, high comedy
        [3, 9],    # Low action, high comedy
        [8, 2],    # High action, low comedy
        [4, 3],    # Low action, low comedy
        [10, 7],   # High action, moderate comedy
        [5, 10]    # Moderate action, high comedy
    ])

    print("User viewing data (action movies, comedy movies):")
    print(user_data)

    # Apply K-means clustering algorithm
    kmeans = KMeans(n_clusters=2, random_state=42)
    kmeans.fit(user_data)

    # Get cluster assignments for each user
    labels = kmeans.labels_

    # Get cluster centers
    centers = kmeans.cluster_centers_

    print("\nCluster centers:")
    print(centers)
    print("\nCluster assignments for each user:")
    print(labels)

    # Visualize the clusters
    plt.figure(figsize=(8, 6))
    plt.scatter(user_data[:, 0], user_data[:, 1], c=labels, cmap='viridis', s=100, alpha=0.8)
    plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200, label='Cluster Centers')
    plt.title('Netflix User Clustering Example')
    plt.xlabel('Action Movies Watched')
    plt.ylabel('Comedy Movies Watched')
    plt.legend()
    plt.grid(True, alpha=0.3)

    # Add annotations to label users
    for i, (x, y) in enumerate(user_data):
        plt.annotate(f'User {i+1}', (x+0.1, y+0.1))

    plt.tight_layout()
    plt.show()

    print("\nInsight: Netflix groups similar users to provide better recommendations")
    print("Users in the same cluster are likely to receive similar content suggestions")
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Create function for Netflix clustering example
def netflix_clustering_example():
    # Create synthetic user viewing data
    # Each row represents a user, columns represent features:
    # [action_movies_watched, comedy_movies_watched]
    user_data = np.array([
        [5, 7],    # Moderate action, more comedy
        [1, 2],    # Little of both
        [8, 8],    # Lots of both
        [12, 10],  # Many action movies
        [2, 3],    # Little of both
        [4, 6],    # Moderate action, comedy
        [10, 9],   # High action, high comedy
        [3, 1],    # Low action, very low comedy
        [7, 5],    # Moderate action, moderate comedy
        [11, 12],  # High action, high comedy
        [2, 8],    # Low action, high comedy
        [9, 3],    # High action, low comedy
        [6, 4],    # Moderate action, low comedy
        [1, 5],    # Very low action, moderate comedy
        [13, 11],  # Very high action, high comedy
        [3, 9],    # Low action, high comedy
        [8, 2],    # High action, low comedy
        [4, 3],    # Low action, low comedy
        [10, 7],   # High action, moderate comedy
        [5, 10]    # Moderate action, high comedy
    ])

    print("User viewing data (action movies, comedy movies):")
    print(user_data)

    # Apply K-means clustering algorithm
    kmeans = KMeans(n_clusters=2, random_state=42)
    kmeans.fit(user_data)

    # Get cluster assignments for each user
    labels = kmeans.labels_

    # Get cluster centers
    centers = kmeans.cluster_centers_

    print("\nCluster centers:")
    print(centers)
    print("\nCluster assignments for each user:")
    print(labels)

    # Create a mesh grid for background coloring
    h = 0.2  # Step size for the mesh
    x_min, x_max = user_data[:, 0].min() - 1, user_data[:, 0].max() + 1
    y_min, y_max = user_data[:, 1].min() - 1, user_data[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    # Predict cluster for each point in the mesh grid
    Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Visualize the clusters with background colors
    plt.figure(figsize=(8, 6))
    # Plot background colors for clusters
    plt.contourf(xx, yy, Z, cmap='viridis', alpha=0.3)
    # Plot user data points
    plt.scatter(user_data[:, 0], user_data[:, 1], c=labels, cmap='viridis', s=100, edgecolors='k', alpha=0.8)
    # Plot cluster centers
    plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200, label='Cluster Centers')
    plt.title('Netflix User Clustering Example')
    plt.xlabel('Action Movies Watched')
    plt.ylabel('Comedy Movies Watched')
    plt.legend()
    plt.grid(True, alpha=0.3)

    # Add annotations to label users
    for i, (x, y) in enumerate(user_data):
        plt.annotate(f'User {i+1}', (x+0.1, y+0.1))

    plt.tight_layout()
    plt.show()

    print("\nInsight: Netflix groups similar users to provide better recommendations")
    print("Users in the same cluster are likely to receive similar content suggestions")
```

```python
# Execute the Netflix example
netflix_clustering_example()
```

##Spotify's Magic with Dimensionality Reduction

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Create function for Spotify PCA example
def spotify_pca_example():
    # Create synthetic song features data
    # Each row represents a song, columns represent audio features:
    # [danceability, energy, acousticness]
    song_data = np.array([
        [2.5, 3.7, 4.5],  # More acoustic, less danceable
        [8.1, 7.2, 5.5],  # Very danceable, high energy
        [1.2, 1.6, 2.8],  # Low energy, less danceable
        [7.3, 8.1, 2.1],  # High energy, very danceable
        [5.5, 4.8, 6.2],  # Moderate across features
        [3.1, 2.9, 5.0],  # Acoustic, low energy
        [9.0, 8.5, 3.0],  # High danceability, high energy
        [2.0, 1.8, 3.5],  # Low danceability, low energy
        [6.5, 7.0, 4.0],  # Moderate danceability, high energy
        [4.8, 5.2, 5.8],  # Moderate across features
        [2.7, 3.0, 4.8],  # Acoustic, low danceability
        [8.5, 7.8, 2.5],  # High danceability, high energy
        [1.5, 2.0, 3.2],  # Low energy, low danceability
        [7.8, 6.5, 3.8],  # High danceability, moderate energy
        [5.0, 4.5, 6.0],  # Moderate across features
        [3.5, 4.0, 5.2],  # Acoustic, moderate energy
        [9.2, 8.0, 2.8],  # High danceability, high energy
        [2.2, 1.5, 4.0],  # Low danceability, high acousticness
        [6.0, 6.8, 4.5],  # Moderate danceability, high energy
        [4.2, 3.8, 5.5]   # Moderate across features
    ])

    song_names = [
        'Calm Acoustic', 'Dance Hit', 'Slow Ballad', 'EDM Track', 'Pop Song',
        'Folk Tune', 'Club Banger', 'Soft Rock', 'Techno Beat', 'Indie Pop',
        'Acoustic Ballad', 'Party Anthem', 'Quiet Melody', 'House Track', 'Chill Pop',
        'Singer-Songwriter', 'Electro Pop', 'Mellow Acoustic', 'Trance Mix', 'Alt Pop'
    ]

    print("Original song features (danceability, energy, acousticness):")
    for i, name in enumerate(song_names):
        print(f"{name}: {song_data[i]}")

    # Apply PCA for dimensionality reduction
    pca = PCA(n_components=2)
    reduced_data = pca.fit_transform(song_data)

    print("\nReduced song features (2 principal components):")
    for i, name in enumerate(song_names):
        print(f"{name}: {reduced_data[i]}")

    print(f"\nExplained variance ratio: {pca.explained_variance_ratio_}")
    print(f"Total variance explained: {sum(pca.explained_variance_ratio_)*100:.2f}%")

    # Apply K-Means clustering to the reduced data for visualization
    kmeans = KMeans(n_clusters=2, random_state=42)
    labels = kmeans.fit_predict(reduced_data)
    centers = kmeans.cluster_centers_

    # Create a mesh grid for background coloring
    h = 0.2  # Step size for the mesh
    x_min, x_max = reduced_data[:, 0].min() - 1, reduced_data[:, 0].max() + 1
    y_min, y_max = reduced_data[:, 1].min() - 1, reduced_data[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    # Predict cluster for each point in the mesh grid
    Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Visualize the PCA results with background colors
    plt.figure(figsize=(10, 7))
    # Plot background colors for clusters
    plt.contourf(xx, yy, Z, cmap='viridis', alpha=0.3)
    # Plot reduced data points
    plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=labels, cmap='viridis', s=100, edgecolors='k', alpha=0.8)
    # Plot cluster centers
    plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200, label='Cluster Centers')
    # Add song name labels
    for i, name in enumerate(song_names):
        plt.annotate(name, (reduced_data[i, 0] + 0.1, reduced_data[i, 1] + 0.1))

    plt.title('Spotify Song Features Reduced with PCA and Clustered')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.legend()
    plt.grid(True, alpha=0.3)

    # Add explanatory arrow and text
    plt.annotate('Similar songs are closer together',
                 xy=(reduced_data[0, 0], reduced_data[0, 1]),
                 xytext=(-5, -3),
                 arrowprops=dict(arrowstyle="->", lw=1.5))

    plt.tight_layout()
    plt.show()

    print("\nInsight: Spotify uses dimensionality reduction to understand")
    print("music similarity in a more compact representation")
```

```python
# Execute the Spotify example
spotify_pca_example()
```
