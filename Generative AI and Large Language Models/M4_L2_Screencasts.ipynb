## M4L2SC1: Text-to-Image Generation with Stable Diffusion

### Step 1: Setting Up Your Environment

```python
# Note: This may take a few minutes
!pip install transformers diffusers Pillow

from diffusers import StableDiffusionPipeline
import torch

model_id = "CompVis/stable-diffusion-v1-4"
device = "cuda" if torch.cuda.is_available() else "cpu"
pipe = StableDiffusionPipeline.from_pretrained(model_id).to(device)
```

### Step 2: Generating Images from Text

```python
prompt = "A serene beach at sunset with gentle waves"
num_inference_steps = 50  # Determines the resolution and clarity
guidance_scale = 7.5  # Adjusts creativity versus prompt fidelity
seed = 42  # Use this to replicate specific outcomes for consistency
generator = torch.manual_seed(seed)
image = pipe(prompt, num_inference_steps=num_inference_steps,
             guidance_scale=guidance_scale, generator=generator).images[0]
```

### Step 3: Displaying and Saving Your Image

```python
image.show()
image.save("generated_image.png")
```

## M4L2SC2: Exploring Latent Space in Diffusion Models

### Step 1: Preparing the Environment

```python
# Note to Ryan: The rest of the notebook can't run without a GPU. (Ryan: üëç)

from diffusers import StableDiffusionPipeline
import torch

model_id = "CompVis/stable-diffusion-v1-4"
pipe = StableDiffusionPipeline.from_pretrained(model_id).to("cuda")

def explore_latent_space(seed):
    torch.manual_seed(seed)
    return pipe.generator
```

### Step 2: Adjusting Latent Variables

```python
seed_values = [42, 1337, 2022]  # Different seeds for variety
results = []

for seed in seed_values:
    torch.manual_seed(seed)
    generator = pipe.generator
    image = pipe(prompt="A futuristic cityscape", generator=generator).images[0]
    results.append(image)
```

### Step 3: Fine-Tuning Outputs

```python
def perturb_latent(image, adjustment_factor):
    # Simulated perturbation function tweaking latent vectors
    perturbed_image = image + adjustment_factor * torch.randn(image.size())
    return perturbed_image

adjustment_factor = 0.1
refined_image = perturb_latent(results[0], adjustment_factor)
```
