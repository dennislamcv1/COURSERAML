### Step 2: Loading and Configuring CLIP

```python
from transformers import CLIPProcessor, CLIPModel
from PIL import Image
import requests

# Load CLIP model and processor
clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch16")
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch16")
clip_model.eval()
```

### Step 3: Performing Multimodal Tasks

```python
# Define the text
text = "An adorable puppy playing in a field"

# Load the image
image_url = "https://images.pexels.com/photos/1458916/pexels-photo-1458916.jpeg?cs=srgb&dl=pexels-poodles2doodles-1458916.jpg&fm=jpg"
image = Image.open(requests.get(image_url, stream=True).raw)

# Process inputs and model inference
inputs = clip_processor(text=[text], images=image, return_tensors="pt", padding=True)

# Calculate similarity
outputs = clip_model(**inputs)
logits_per_image = outputs.logits_per_image
probs = logits_per_image.softmax(dim=1)

# Output the similarity probabilities
print("Similarity probabilities:", probs)
```
